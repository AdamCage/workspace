# Отток клиентов

Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Вам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком. 

Постройте модель с предельно большим значением *F1*-меры. Чтобы сдать проект успешно, нужно довести метрику до 0.59. Проверьте *F1*-меру на тестовой выборке самостоятельно.

Дополнительно измеряйте *AUC-ROC*, сравнивайте её значение с *F1*-мерой.

Источник данных: [https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling](https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling)

Цель данной работы - построить модель, предсказывающую возможность уход клиента из банка в ближайшее время с метрикой `F1-мера` не менее 0.59.

Работа пройдет в 5 этапов:
1. Подготовка данных;
2. Исследование задачи;
3. Борьба с дисбалансом;
4. Тестирование модели;
5. Вывод.

Признаки
- RowNumber — индекс строки в данных
- CustomerId — уникальный идентификатор клиента
- Surname — фамилия
- CreditScore — кредитный рейтинг
- Geography — страна проживания
- Gender — пол
- Age — возраст
- Tenure — сколько лет человек является клиентом банка
- Balance — баланс на счёте
- NumOfProducts — количество продуктов банка, используемых клиентом
- HasCrCard — наличие кредитной карты
- IsActiveMember — активность клиента
- EstimatedSalary — предполагаемая зарплата

Целевой признак

#### Препроцессинг данных

Проведем препроцессинг следующим образом:
- Преобразуем категориальные параметры методом порядкового кодирования с последующей заменой категорий вероятностью `target` для этой категории.

```python
%%time
### ------------------- Раскомментировать код для использования других способов препроцессинга

# st_scaler = StandardScaler()
# min_max_scaler = MinMaxScaler()

encoder = OrdinalEncoder()

for i in ds.drop('target', axis=1):
    if not pd.isna(re.search('^c.*', i)):
        ds[i] = pd.DataFrame(encoder.fit_transform(ds[i].reset_index()),
                             columns=ds[i].reset_index().columns).drop('index', axis=1)
        ds[i] = ds[i].apply(
            lambda x: round(ds.query(f'{i} == 1')['target'].mean(), 4) if x == 1 else round(
                ds.query(f'{i} == 0')['target'].mean(), 4
            )
        )

# X_train = pd.get_dummies(X_train, drop_first=True)

ds.sample(5)
```

# Исследование задачи

Учитывая несбалансированность целевого признака, построим модели на текущем датасете и оценим их результат.

Для это будем использовать следующую последовательность действий:
1. Определим фичи `features` и таргет `target`;
2. Сформируем обучающую, валидационную и тестовую выборку с соотношением 60:20:20 (3:1:1) соответственно;
3. Построим модели, определим наилучшие гиперпараметры для них с помощью метода `greed_search` и обучим их на обучающей выборке и проверим их результат;
4. Оценим эффективность моделей на валидационной выборке для уменьшения вероятности переобучения.

В качестве моделей используем `логистическую регрессию`, `дерево решений` и `случайный лес`.

## Определение фичей и таргета

```python
features = ds.drop('target', axis=1)
target = ds['target']
```

## Формирование выборок

```python
# test - 20%, train & validation - 80%
X_other, X_test, y_other, y_test = train_test_split(features, target, test_size=0.2, random_state=RANDOM_STATE, stratify=target)

# train - 60%, validation - 20%
X_train, X_val, y_train, y_val = train_test_split(X_other, y_other, test_size=0.25, random_state=RANDOM_STATE, stratify=y_other)

print(f'Размер обучающей выборки: {X_train.shape[0]}')
print(f'Размер валидационной выборки: {X_val.shape[0]}')
print(f'Размер тестовой выборки: {X_test.shape[0]}')
```

#### Нормализация выборки

Проведем препроцессинг следующим образом:
1. Нормализуем количественные параметры с помощью метода min-max, так как не все параметры распределены нормально: в данных имеются выбросы, а также значительная концентрация 0 в параметре `q_balance`;
2. Преобразуем категориальные параметры методом порядкового кодирования с последующей заменой категорий вероятностью `target` для этой категории.

```python
%%time
### ------------------- Раскомментировать код для использования других способов препроцессинга

# st_scaler = StandardScaler()
min_max_scaler = MinMaxScaler()


X_train = X_train.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)

X_val = X_val.reset_index(drop=True)
y_val = y_val.reset_index(drop=True)

X_test = X_test.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)

        
min_max_scaler.fit(X_train)

X_train = pd.DataFrame(
    min_max_scaler.transform(X_train),
    columns=X_train.columns
)

X_val = pd.DataFrame(
    min_max_scaler.transform(X_val),
    columns=X_val.columns
)

X_test = pd.DataFrame(
    min_max_scaler.transform(X_test),
    columns=X_test.columns
)


X_train.sample(5)
```
